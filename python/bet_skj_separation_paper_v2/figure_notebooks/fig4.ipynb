```python
# - Define constant plot params
stipsizenow = 10
stipmarknow = 'o'
stipfacecolnow = 'none'
stipedgeltcolnow = 'lightgray'
stipedgedkcolnow = 'darkslategray'
stipewnow = 0.8 # marker edge width
cmnow = cm.PuOr
extendnow = 'both'
```

```python
# - Define subplot-variable plot params
mapsnow = [skj_cp_tot_anom_en, bet_cp_tot_anom_en,
           bettoskj_cp_tot_anom_en,
           skj_cp_tot_anom_ln, bet_cp_tot_anom_ln,
           bettoskj_cp_tot_anom_ln]
pvsnow = [skj_cp_tot_anom_enln_wrs_pval,
          bet_cp_tot_anom_enln_wrs_pval,
          bettoskj_cp_tot_anom_enln_wrs_pval,
          skj_cp_tot_anom_enln_wrs_pval,
          bet_cp_tot_anom_enln_wrs_pval,
          bettoskj_cp_tot_anom_enln_wrs_pval]
ptfsnow = [skj_cp_tot_anom_enln_wrs_ptf,
           bet_cp_tot_anom_enln_wrs_ptf,
           bettoskj_cp_tot_anom_enln_wrs_ptf,
           skj_cp_tot_anom_enln_wrs_ptf,
           bet_cp_tot_anom_enln_wrs_ptf,
           bettoskj_cp_tot_anom_enln_wrs_ptf]
vmaxsnow = [500, 50, 0.08, 500, 50, 0.08]
vminsnow = -1*np.asarray(vmaxsnow)
stipltdkcosnow = 0.5*np.asarray(vmaxsnow) # light/dark stip cutoff value
titlesnow = ['SKJ CPUE - El Niño anom', 'BET CPUE - El Niño anom',
             'BET:SKJ CPUE - El Niño anom',
             'SKJ CPUE - La Niña anom', 'BET CPUE - La Niña anom',
             'BET:SKJ CPUE - La Niña anom']
unitsnow = ['[metric tons/set]', '[metric tons/set]', '',
            '[metric tons/set]', '[metric tons/set]', '']
signifstipsnow = [1,1,1,1,1,1]
fishareastipsnow = [1,1,1,1,1,1]
ploteezsnow = [0,0,0,0,0,0]
```

```python
# - Define pts where fish area criteria are met
fisharea_ennow = (
    (skj_cp_tot_anom_en>=0).values & \
    (bet_cp_tot_anom_en<=0).values & \
    (bettoskj_cp_tot_anom_en<=0).values)
[skjsigniflon_ennow,skjsigniflat_ennow] = \
    find_where_pval_small(
        skj_cp_tot_anom_enln_wrs_pval.where(fisharea_ennow),
        skj_cp_tot_anom_enln_wrs_ptf) 
[betsigniflon_ennow,betsigniflat_ennow] = \
    find_where_pval_small(
        bet_cp_tot_anom_enln_wrs_pval.where(fisharea_ennow),
        bet_cp_tot_anom_enln_wrs_ptf) 
[bettoskjsigniflon_ennow,bettoskjsigniflat_ennow] = \
    find_where_pval_small(
        bettoskj_cp_tot_anom_enln_wrs_pval.where(fisharea_ennow),
        bettoskj_cp_tot_anom_enln_wrs_ptf) 
    
fisharea_lnnow = (
    (skj_cp_tot_anom_ln>=0).values & \
    (bet_cp_tot_anom_ln<=0).values & \
    (bettoskj_cp_tot_anom_ln<=0).values)
[skjsigniflon_lnnow,skjsigniflat_lnnow] = \
    find_where_pval_small(
        skj_cp_tot_anom_enln_wrs_pval.where(fisharea_lnnow),
        skj_cp_tot_anom_enln_wrs_ptf) 
[betsigniflon_lnnow,betsigniflat_lnnow] = \
    find_where_pval_small(
        bet_cp_tot_anom_enln_wrs_pval.where(fisharea_lnnow),
        bet_cp_tot_anom_enln_wrs_ptf) 
[bettoskjsigniflon_lnnow,bettoskjsigniflat_lnnow] = \
    find_where_pval_small(
        bettoskj_cp_tot_anom_enln_wrs_pval.where(fisharea_lnnow),
        bettoskj_cp_tot_anom_enln_wrs_ptf) 
```

```python
# - Set proj and define axes
fig,axes = plt.subplots(nrows=2, ncols=3, figsize=(16,7.5),
                        subplot_kw={'projection': ccrs.PlateCarree(central_longitude=200)})

# - Make maps pretty + plot
isp = 0
for irow in range(axes.shape[0]):
    for icol in range(axes.shape[1]):
        ax = axes[irow][icol]
        exec(open('helper_scripts/create_map_bgs.py').read())
        
        mapsnow[isp].plot(
            ax=ax, transform=ccrs.PlateCarree(), cmap=cmnow,
            vmin=vminsnow[isp], vmax=vmaxsnow[isp],
            cbar_kwargs={'pad': 0.02, 'label': unitsnow[isp],
                         'extend': extendnow})
        
        if ploteezsnow[isp]==1:
            nueezs.plot(ax=ax, transform=ccrs.PlateCarree(),
                color=eezfcnow, edgecolor=eezlcnow, linewidth=eezlwnow)
            
        if fishareastipsnow[isp]==1:
            if irow==0: # El Nino
                fishareanow = fisharea_ennow
                skjsigniflonnow = skjsigniflon_ennow
                skjsigniflatnow = skjsigniflat_ennow
                betsigniflonnow = betsigniflon_ennow
                betsigniflatnow = betsigniflat_ennow
                bettoskjsigniflonnow = bettoskjsigniflon_ennow
                bettoskjsigniflatnow = bettoskjsigniflat_ennow
            elif irow==1: # La Nina
                fishareanow = fisharea_lnnow
                skjsigniflonnow = skjsigniflon_lnnow
                skjsigniflatnow = skjsigniflat_lnnow
                betsigniflonnow = betsigniflon_lnnow
                betsigniflatnow = betsigniflat_lnnow
                bettoskjsigniflonnow = bettoskjsigniflon_lnnow
                bettoskjsigniflatnow = bettoskjsigniflat_lnnow
                
            # - Find spatial pts where fish area criteria are met
            fishareamapnow = skj_cp_tot_anom_en.where(fishareanow)
            # ...doesn't matter which underlying map is used,
            # just want the lat/lon associated w/ any WCPFC gridded map
            
            # - Get lat,lon where fish area criteria are met
            fishareastackednow = fishareamapnow.stack(x=['lat','lon'])
            fishareastackednow = fishareastackednow[fishareastackednow.notnull()]
            fisharealonnow = fishareastackednow.lon
            fisharealatnow = fishareastackednow.lat
            
            # - Plot pts where fish area criteria are met
            #ax.scatter(fisharealonnow, fisharealatnow,
            #        c='magenta', s=10, transform=ccrs.PlateCarree());
            
            # - Plot pts where fish area criteria are met AND at
            # least one map a shows signif diff btwn EN/LN
            skjsigniflistnow = list(zip(skjsigniflonnow.values,
                        skjsigniflatnow.values))
            betsigniflistnow = list(zip(betsigniflonnow.values,
                                        betsigniflatnow.values))
            bettoskjsigniflistnow = list(zip(bettoskjsigniflonnow.values,
                                             bettoskjsigniflatnow.values))
            signiflistnow = set(skjsigniflistnow) | \
                            set(betsigniflistnow) | \
                            set(bettoskjsigniflistnow)
            if len(signiflistnow)>0:
                signiflonnow, signiflatnow = zip(*signiflistnow)
                ax.scatter(signiflonnow, signiflatnow,
                    facecolors='none', edgecolors='darkslategray',
                    marker='s', s=40, linewidths=1,
                    transform=ccrs.PlateCarree())
                
        if signifstipsnow[isp]==1:
            [ltcol_signiflonnow,ltcol_signiflatnow]=find_where_pval_small(
                pvsnow[isp].where(abs(mapsnow[isp])>stipltdkcosnow[isp]),
                ptfsnow[isp])
            [dkcol_signiflonnow,dkcol_signiflatnow]=find_where_pval_small(
                pvsnow[isp].where(abs(mapsnow[isp])<=stipltdkcosnow[isp]),
                ptfsnow[isp])
            ax.scatter(ltcol_signiflonnow, ltcol_signiflatnow,
                       marker=stipmarknow, linewidths=stipewnow,
                       facecolors=stipfacecolnow, edgecolors=stipedgeltcolnow,
                       s=stipsizenow, transform=ccrs.PlateCarree())
            ax.scatter(dkcol_signiflonnow, dkcol_signiflatnow,
                       marker=stipmarknow, linewidths=stipewnow,
                       facecolors=stipfacecolnow, edgecolors=stipedgedkcolnow,
                       s=stipsizenow, transform=ccrs.PlateCarree())
        
                    
        ax.set_xlabel(''); ax.set_ylabel('')
        ax.set_title(titlesnow[isp])
        isp = isp + 1
       
fig.savefig(figpath + 'mainfig4_bscpue_ensomeans.png', bbox_inches='tight', pad_inches = 0, dpi = 200)
```
