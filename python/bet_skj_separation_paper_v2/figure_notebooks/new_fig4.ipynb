```python
# - Define subplot-variable plot params
if (totanomnow=='totmmean') | (totanomnow=='anom'):
    cmsnow = 9*[cm.PuOr]
    stipedgedkcolsnow = 9*['darkslategray']
    signifstipsnow = 9*[1]
    ploteezsnow = 9*[0]
    if totanomnow=='anom':
        vmaxsnow = [10, 3.5, 0.08, 10, 3.5, 0.08, 18, 6.5, 0.1]
        mapsnow = [skj_cp_tot_anom_en, bet_cp_tot_anom_en,
                   bettoskj_cp_tot_anom_en,
                   skj_cp_tot_anom_ln, bet_cp_tot_anom_ln,
                   bettoskj_cp_tot_anom_ln,
                   skj_cp_tot_anom_en-skj_cp_tot_anom_ln,
                   bet_cp_tot_anom_en-bet_cp_tot_anom_ln,
                   bettoskj_cp_tot_anom_en-bettoskj_cp_tot_anom_ln]
        pvsnow = 3*[skj_cp_tot_anom_enln_wrs_pval,
                  bet_cp_tot_anom_enln_wrs_pval,
                  bettoskj_cp_tot_anom_enln_wrs_pval]
        ptfsnow = 3*[skj_cp_tot_anom_enln_wrs_ptf,
                   bet_cp_tot_anom_enln_wrs_ptf,
                   bettoskj_cp_tot_anom_enln_wrs_ptf]
        titlesnow = ['SKJ CPUE - El Niño anom', 'BET CPUE - El Niño anom',
                     'BET:SKJ CPUE - El Niño anom',
                     'SKJ CPUE - La Niña anom', 'BET CPUE - La Niña anom',
                     'BET:SKJ CPUE - La Niña anom',
                     'SKJ CPUE - EN-LN anom', 'BET CPUE - EN-LN anom',
                     'BET:SKJ CPUE - EN-LN anom']
    elif totanomnow=='totmmean':
        vmaxsnow = [12, 4, 0.12, 12, 4, 0.12, 18, 10, 0.16]
        mapsnow = [skj_cp_tot_en-skj_cp_tot_mean,
                   bet_cp_tot_en-bet_cp_tot_mean,
                   bettoskj_cp_tot_en-bettoskj_cp_tot_mean,
                   skj_cp_tot_ln-skj_cp_tot_mean,
                   bet_cp_tot_ln-bet_cp_tot_mean,
                   bettoskj_cp_tot_ln-bettoskj_cp_tot_mean,
                   skj_cp_tot_en-skj_cp_tot_ln,
                   bet_cp_tot_en-bet_cp_tot_ln,
                   bettoskj_cp_tot_en-bettoskj_cp_tot_ln]
        pvsnow = 3*[skj_cp_tot_enln_wrs_pval,
                  bet_cp_tot_enln_wrs_pval,
                  bettoskj_cp_tot_enln_wrs_pval]
        ptfsnow = 3*[skj_cp_tot_enln_wrs_ptf,
                   bet_cp_tot_enln_wrs_ptf,
                   bettoskj_cp_tot_enln_wrs_ptf]
        titlesnow = ['SKJ CPUE - El Niño minus mean', 'BET CPUE - El Niño minus mean',
                     'BET:SKJ CPUE - El Niño minus mean',
                     'SKJ CPUE - La Niña minus mean', 'BET CPUE - La Niña minus mean',
                     'BET:SKJ CPUE - La Niña minus mean',
                     'SKJ CPUE - EN-LN', 'BET CPUE - EN-LN',
                     'BET:SKJ CPUE - EN-LN']
    vminsnow = -1*np.asarray(vmaxsnow)
elif totanomnow=='tot':
    cmsnow = [cm.viridis, cm.viridis, cm.viridis,
              cm.viridis, cm.viridis, cm.viridis,
              cm.PuOr, cm.PuOr, cm.PuOr]
    stipedgedkcolsnow = ['lightgray','lightgray','lightgray',
                         'lightgray','lightgray','lightgray',
                         'darkslategray','darkslategray','darkslategray']
    signifstipsnow = 9*[0,0,0,0,0,0,1,1,1]
    ploteezsnow = 9*[1,1,1,1,1,1,0,0,0]
    vmaxsnow = [60, 15, 0.32, 60, 15, 0.32, 18, 10, 0.16]
    vminsnow = [0, 0, 0, 0, 0, 0, -18, -10, -0.16]
    mapsnow = [skj_cp_tot_en, bet_cp_tot_en,
               bettoskj_cp_tot_en,
               skj_cp_tot_ln, bet_cp_tot_ln,
               bettoskj_cp_tot_ln,
               skj_cp_tot_en-skj_cp_tot_ln,
               bet_cp_tot_en-bet_cp_tot_ln,
               bettoskj_cp_tot_en-bettoskj_cp_tot_ln]
    pvsnow = 3*[skj_cp_tot_enln_wrs_pval,
              bet_cp_tot_enln_wrs_pval,
              bettoskj_cp_tot_enln_wrs_pval]
    ptfsnow = 3*[skj_cp_tot_enln_wrs_ptf,
               bet_cp_tot_enln_wrs_ptf,
               bettoskj_cp_tot_enln_wrs_ptf]
    titlesnow = ['SKJ CPUE - El Niño', 'BET CPUE - El Niño',
                 'BET:SKJ CPUE - El Niño',
                 'SKJ CPUE - La Niña', 'BET CPUE - La Niña',
                 'BET:SKJ CPUE - La Niña',
                 'SKJ CPUE - EN-LN', 'BET CPUE - EN-LN',
                 'BET:SKJ CPUE - EN-LN']

stipltdkcosnow = 0.5*np.asarray(vmaxsnow) # light/dark stip cutoff value
unitsnow = 3*['[metric tons/set]', '[metric tons/set]', '']
```
