```python
lightgray = np.array([236/256, 236/256, 236/256, 1])
blue = np.array([0, 0, 256/256, 1])
red = np.array([256/256, 0, 0, 1])
green = np.array([0, 256/256, 0, 1])
orange = np.array([255/256, 140/256, 0, 1])
newcolors = np.vstack((lightgray,blue,red,green,orange))
bs5cmp = ListedColormap(newcolors)
```

```python
import matplotlib.gridspec as gridspec
```

```python
def plotfdbpmap_hatch(fdmapnow,fdmapstrnow):
    dsnow = bettoskj_cp_tot_mean.to_dataset(name='bettoskj_cp_tot_mean')
    dsnow['skj_cp_tot_mean'] = skj_cp_tot_mean
    dsnow['bet_cp_tot_mean'] = bet_cp_tot_mean
    dsnow['bettoskj_cp_tot_mean'] = bettoskj_cp_tot_mean
    dsnow['fdmapnow'] = fdmapnow
    dfnow = dsnow.to_dataframe()
    fig = plt.figure(constrained_layout=True, figsize=(9,5.5))
    gs = fig.add_gridspec(3, 12)
    ax1 = fig.add_subplot(gs[0,0:3])
    ax2 = fig.add_subplot(gs[0,4:7])
    ax3 = fig.add_subplot(gs[0,8:11])
    dfnow.boxplot(ax=ax1, column='skj_cp_tot_mean', by='fdmapnow'); ax1.set_xlabel('')
    dfnow.boxplot(ax=ax2, column='bet_cp_tot_mean', by='fdmapnow'); ax2.set_xlabel('')
    dfnow.boxplot(ax=ax3, column='bettoskj_cp_tot_mean', by='fdmapnow'); ax3.set_xlabel('')
    ax4 = fig.add_subplot(gs[1:,:], projection = ccrs.PlateCarree(central_longitude=200))
    ax = ax4; exec(open('helper_scripts/create_map_bgs.py').read())
    fdmapnow_hatch = xr.where(fdmapnow % 1 == 0, fdmapnow, np.nan)  
    pltnow = fdmapnow.plot(ax=ax4, add_colorbar=False, transform=ccrs.PlateCarree(), cmap=bs5cmp, vmin=-0.5, vmax=4.5) #, add_guide=None)
    ax4.pcolor(fdmapnow_hatch.lon-2.5, fdmapnow_hatch.lat-2.5, fdmapnow_hatch, hatch='////', alpha=0, transform=ccrs.PlateCarree())
    ax4.set_title('')
    cbar = plt.colorbar(pltnow, ax=ax4, ticks=np.arange(-0.5,4.5,5/5) + 5/5/2)
    cbar.ax.set_yticklabels(['None','LS-LB','LS-HB','HS-LB','HS-HB']);
    eezfcnow = 'none'; eezlcnow = 'black'; eezlsnow = '-'; eezlwnow = 1
    nueezs.plot(ax=ax4, transform=ccrs.PlateCarree(), color=eezfcnow, edgecolor=eezlcnow, linewidth=eezlwnow, linestyle=eezlsnow)
    fig.savefig(figpath + fdmapstrnow + '.png',
                bbox_inches='tight', pad_inches = 0, dpi = 200)
```
