```python
# - Define constant plot params
cmnow = cm.PuOr

signifstipnow = 1
stipsizenow = 10
stipmarknow = 'o'
stipfacecolnow = 'none'
stipltdkconow = 0.45 # light/dark Pearson CC cutoff
stipedgeltcolnow = 'lightgray'
stipedgedkcolnow = 'darkslategray'
stipewnow = 0.8 # marker edge width

separeastipnow = 1
```

```python
# - Plot all env vars except SSHA (9 total),
# conveniently the numerically last one in list

nrows = 3; ncols = 3
for ifig in range(0,len(bsvsenv_all_maps_cc)):
    
    # - Set proj and define axes
    fig,axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(16,6.5),
        subplot_kw={'projection': ccrs.PlateCarree(central_longitude=200)})
    
    # - Make maps pretty
    for _, ax in enumerate(fig.axes):
        exec(open('helper_scripts/create_map_bgs.py').read())
    
    iiv = 0
    for yax in range(0,nrows):
        for xax in range(0,ncols):
            ax = axes[yax][xax]
            
            # - Plot correl + pval
            bsvsenv_all_maps_cc[ifig][iiv].plot(
                ax=ax, transform=ccrs.PlateCarree(),
                cmap=cmnow, vmin=-1, vmax=1,
                cbar_kwargs={'pad': 0.02, 'label': 'Pearson\'s R'});
            
            if separeastipnow==1:
                # - Plot pts where separation area criteria are met
                #ax.scatter(separealon_ev[iiv], separealat_ev[iiv],
                #    c='magenta', s=stipsizenow,
                #    transform=ccrs.PlateCarree())
                
                # - Plot pts where separation area criteria are met AND at
                # least 1 map a shows signif correl w/ the given env var 
                skjsigniflistnow = list(zip(skjsigniflon_ev[iiv].values,
                                            skjsigniflat_ev[iiv].values))
                betsigniflistnow = list(zip(betsigniflon_ev[iiv].values,
                                            betsigniflat_ev[iiv].values))
                bettoskjsigniflistnow = list(zip(bettoskjsigniflon_ev[iiv].values,
                                                 bettoskjsigniflat_ev[iiv].values))
                signiflistnow = set(skjsigniflistnow) | \
                                set(betsigniflistnow) | \
                                set(bettoskjsigniflistnow)
                if len(signiflistnow)>0:
                    signiflonnow, signiflatnow = zip(*signiflistnow)
                    ax.scatter(signiflonnow, signiflatnow,
                        facecolors='none', edgecolors='darkslategray',
                        marker='s', s=40, linewidths=1,
                        transform=ccrs.PlateCarree())
            
            if signifstipnow==1:
                [ltcol_signiflonnow,ltcol_signiflatnow]=find_where_pval_small(
                    bsvsenv_all_maps_cc_pval[ifig][iiv].where(
                        abs(bsvsenv_all_maps_cc[ifig][iiv])>stipltdkconow),
                    bsvsenv_all_maps_cc_ptf[ifig][iiv])
                [dkcol_signiflonnow,dkcol_signiflatnow]=find_where_pval_small(
                    bsvsenv_all_maps_cc_pval[ifig][iiv].where(
                        abs(bsvsenv_all_maps_cc[ifig][iiv])<=stipltdkconow),
                    bsvsenv_all_maps_cc_ptf[ifig][iiv])
                ax.scatter(ltcol_signiflonnow, ltcol_signiflatnow,
                    marker=stipmarknow, linewidths=stipewnow,
                    facecolors=stipfacecolnow, edgecolors=stipedgeltcolnow,
                    s=stipsizenow, transform=ccrs.PlateCarree())
                ax.scatter(dkcol_signiflonnow, dkcol_signiflatnow,
                    marker=stipmarknow, linewidths=stipewnow,
                    facecolors=stipfacecolnow, edgecolors=stipedgedkcolnow,
                    s=stipsizenow, transform=ccrs.PlateCarree())
            
            ax.set_xlabel(''); ax.set_ylabel('');
            ax.set_title(bsvsenv_dvnicenamelist[ifig] + ' vs. ' + 
                bsvsenv_ivnicenamelist[iiv] + ' correlation')
            
            iiv = iiv + 1
            
    fig.savefig(figpath + 'suppfig' + str(ifig+11) + '_' +
                bsvsenv_dvnamelist[ifig] + 'vsenv_cc_maps.png',
                bbox_inches='tight', pad_inches = 0, dpi = 200)
```
