## Look for correlation btwn predictors 

```python
def calculate_pvalues(df):
    df = df._get_numeric_data()
    dfcols = pd.DataFrame(columns=df.columns)
    pvalues = dfcols.transpose().join(dfcols, how='outer')
    #print(pvalues)
    for r in df.columns:
        for c in df.columns:
            #print(r,c)
            if c == r:
                pvalues[r][c] = 0
            else:
                df_corr = df[[r,c]].dropna()
                if len(df_corr[r])>2 and len(df_corr[c])>2:
                    pvalues[r][c] = pearsonr(df_corr[r], df_corr[c])[1]
    return pvalues
```

```python
axes = pd.plotting.scatter_matrix(df, alpha = 0.5, figsize = (15,15), diagonal='hist');
corrnow = df.corr().as_matrix()
pvalnow = calculate_pvalues(df).as_matrix()
for i, j in zip(*plt.np.triu_indices_from(axes, k=1)):
    colnow = 'red' if corrnow[i,j]**2 > 0.6 else 'black'
    axes[i, j].annotate("R^2=%.3f" %corrnow[i,j]**2, (0.5, 0.5), 
                        xycoords='axes fraction', ha='center', va='center', color=colnow)
    axes[i, j].annotate("%.2f" %pvalnow[i,j], (0.5, 0.3), 
                        xycoords='axes fraction', ha='center', va='center')
fig.savefig(figpath + 'question3_spmatrix_IVs.png', bbox_inches='tight', pad_inches = 0, dpi = 200)
```

Collinearity btwn environmental variables is high when R^2 > abs(0.6) (Arrizabalaga et al., 2015).  
So there's high collinearity between sst and temp100 (R^2 = 0.842), sst and o2100 (R^2 = 0.655), and temp100 and o2100 (R^2 = 0.64). Also between sst and temp100avg (R^2 = 0.953) (not shown) and temp100avg and o2100 (R^2 = 0.695) (not shown).

## Univariate GAMs for BET:SKJ CPUE ratio

- Use a log transformation on CPUE ratio to better normalize the
assymmetric distribution w/ a factor of 0.1 added before
log-transformation to acct for zero CPUE ratio 
- Degree of smoothness of model terms (env variables) was restricted to 4
- Final models then validated by verifying the residuals of the GAMs for
normality, homogeneity (relationship btwn residuals against the fitted values?),
independence, over-dispersion parameter for over-dispersion of residuals (Hu et al., 2018)
- Unlike the Quotient Analysis, in which preference was defined as the ratio of the
fraction of total CPUE to the fraction of area significantly higher than 1,
the definition of the preference in GAMs was simply based upon higher CPUE.
- LinearGAM: identity link and normal distribution

```python
#https://pygam.readthedocs.io/en/latest/api/lineargam.html
from pygam import LinearGAM, s
```

```python
# - Hu et al. (2018)
#Table 4 shows the explained deviance in the univariate GAMs for the pre-FAD (1981–1993)
#and FAD time periods (1994–2012).

# - Arrizabalaga et al. (2015)
#...The deviance explained by the model was estimated as 1-(residual deviance)/(null deviance),
#and is the equivalent to R2 in regression models (Guisan and Zimmermann, 2000).
#...In the univariate GAMs, the “stock” factor explained a relatively high proportion of the null deviance
#in the case of skipjack (23%), bigeye (20%) and albacore (12%), but less than 1% for other species (Table 1).
#The percent deviance explained by the “year” factor varied between 2% (in the case of bigeye) and 21%
#(in the case of skipjack), while that explained by the “season” factor was minor ( o 2% in all cases except for bluefin).
#TEMP5 was the most important variable for yellowfin tuna, as TEMP100 was for south- ern bluefin.
#The proportion of null deviance explained by these two temperature related variables ranged between 38%
#(for south- ern bluefin) and 2% (skipjack) of total deviance. Oxygen, highly correlated with temperature,
#explained similar proportions. Sali- nity explained between 17% and 4% of the null deviance for southern bluefin,
#skipjack, albacore and Atlantic bluefin, with less than 2% for yellowfin and bigeye.

# - pygam documentation (pg. 115):
#deviance(y, mu, scaled=True)
#model deviance
#for a gaussian linear model, this is equal to the SSE
#Parameters
#• y (array-like of length n) – target values
#• mu (array-like of length n) – expected values
#• scaled (boolean, default: True) – whether to divide the deviance by the dis- tribution scaled
#Returns deviances
#Return type np.array of length n
```

```python
df_uni_gam_lb2scptot = pd.DataFrame(columns=['Variable', 'Explained Deviance', 'AIC', 'AICc', 'GCV', 'lambda'])
for var in df.columns[3:-3]:
    dfnow = df[['lb2scptot',var]].dropna()
    gam = LinearGAM(s(0)).fit(dfnow[var], dfnow['lb2scptot'])
    lam = np.logspace(-3, 5, 5)
    gam.gridsearch(dfnow[var], dfnow['lb2scptot'], lam=lam)
    df_uni_gam_lb2scptot = df_uni_gam_lb2scptot.append(
        {'Variable': var, 'Explained Deviance': gam.statistics_['pseudo_r2']['explained_deviance'],
         'AIC': gam.statistics_['AIC'], 'AICc': gam.statistics_['AICc'], 'GCV': gam.statistics_['GCV'], 
         'lambda': gam.lam[0][0]},
        ignore_index=True)
df_uni_gam_lb2scptot
```

## GAMS Example 2:

```python
from pygam.datasets import wage
X, y = wage()
gam = LinearGAM(s(0) + s(1) + f(2)).fit(X, y)
gam.summary()
```

```python
lam = np.logspace(-3, 5, 5)
lams = [lam] * 3
gam.gridsearch(X, y, lam=lams)
gam.summary()
```

```python
for i, term in enumerate(gam.terms):
    if term.isintercept:
        continue
    XX = gam.generate_X_grid(term=i)
    pdep, confi = gam.partial_dependence(term=i, X=XX, width=0.95)
    plt.figure()
    plt.plot(XX[:, term.feature], pdep)
    plt.plot(XX[:, term.feature], confi, c='r', ls='--')
    plt.title(repr(term))
```
