## Look for correlation btwn predictors 

```python
def calculate_pvalues(df):
    df = df._get_numeric_data()
    dfcols = pd.DataFrame(columns=df.columns)
    pvalues = dfcols.transpose().join(dfcols, how='outer')
    #print(pvalues)
    for r in df.columns:
        for c in df.columns:
            #print(r,c)
            if c == r:
                pvalues[r][c] = 0
            else:
                df_corr = df[[r,c]].dropna()
                if len(df_corr[r])>2 and len(df_corr[c])>2:
                    pvalues[r][c] = pearsonr(df_corr[r], df_corr[c])[1]
    return pvalues
```

```python
#%%script false --no-raise-error
# include the line above if you don't want to run this cell

axes = pd.plotting.scatter_matrix(df, alpha = 0.5, figsize = (15,15), diagonal='hist');
corrnow = df.corr().as_matrix()
pvalnow = calculate_pvalues(df).as_matrix()
for i, j in zip(*plt.np.triu_indices_from(axes, k=1)):
    colnow = 'red' if corrnow[i,j]**2 > 0.6 else 'black'
    axes[i, j].annotate("R^2=%.3f" %corrnow[i,j]**2, (0.5, 0.5), 
                        xycoords='axes fraction', ha='center', va='center', color=colnow)
    axes[i, j].annotate("%.2f" %pvalnow[i,j], (0.5, 0.3), 
                        xycoords='axes fraction', ha='center', va='center')
fig.savefig(figpath + 'question3_spmatrix_IVs.png', bbox_inches='tight', pad_inches = 0, dpi = 200)
```

### 1a.) Univariate GAMs for BET:SKJ CPUE ratio w/ log(CPUE ratio) + normal dist + identity link

```python
df_uni_gam_lb2scptot = pd.DataFrame(columns=['Variable', 'Explained Deviance', 'AIC', 'AICc', 'GCV', 'lambda'])
fig,axes = plt.subplots(nrows=2, ncols=5, figsize=(16,6))
titles = df.columns[3:-3].values
iplot = 0
for var in df.columns[3:-3]:
    dfnow = df[['lb2scptot',var]].dropna()
    gam = LinearGAM(s(0)).fit(dfnow[var], dfnow['lb2scptot'])
    lam = np.logspace(-3, 5, 5)
    gam.gridsearch(dfnow[var], dfnow['lb2scptot'], lam=lam)
    df_uni_gam_lb2scptot = df_uni_gam_lb2scptot.append(
        {'Variable': var, 'Explained Deviance': gam.statistics_['pseudo_r2']['explained_deviance'],
         'AIC': gam.statistics_['AIC'], 'AICc': gam.statistics_['AICc'], 'GCV': gam.statistics_['GCV'], 
         'lambda': gam.lam[0][0]},
        ignore_index=True)
    # - partial dependence plot
    irow = 0 if iplot < 5 else 1
    icol = iplot if iplot < 5 else iplot-5
    ax = axes[irow][icol]
    XX = gam.generate_X_grid(term=0, n=500)
    ax.plot(XX[:,0], np.zeros(len(XX[:,0])), c='k', ls='--')
    ax.plot(XX[:,0], gam.partial_dependence(term=0, X=XX))
    ax.plot(XX[:,0], gam.partial_dependence(term=0, X=XX, width=.95)[1], c='r', ls='--')
    ax.set_title(titles[iplot])
    iplot = iplot + 1
#fig.savefig(figpath + 'question3_unigamsmooths_lnb2scptot.png', bbox_inches='tight', pad_inches = 0, dpi = 200)
df_uni_gam_lb2scptot
```

**Notes:**
- Hu et al. (2018):  
"Since the CPUE follows a continuous distribution, we assumed the Gaussian distribution, which was associated with the identity link function for the CPUE. We also used a logarithmic transformation on the CPUE to normalize the asymmetrical distribution (Arrizabalaga et al., 2015; Mugo, Saitoh, Nihira, & Kuroyama, 2010). A factor of 0.1 was added before log-transformation to account for zero CPUE (Mugo et al., 2010). The predictor function of our GAM includes environmental variables or covariates and fixed factors to adjust the temporal effects (year, month and their interaction). The degree of smoothness of model terms (i.e., environmental variables) was restricted to 4 (Arrizabalaga et al., 2015)."  
"We assembled these models by first checking for collinearity between covariates; these relationships are shown in Figure 2. Among the covariates only SSChl vs O2[150] during the FAD period (1994–2012) for the longline fishery exhibited collinearity (r = .69), so we excluded SSChl from the analysis. We then assembled a model consisting of all environmental variables and calculated its Akaike Information Criteria (AIC). Akaike Information Criteria (AIC) values were then calculated for models in which each of the environmental variables was individually removed. If the removal of a variable did not lead to a significant increase in AIC, it was dropped from the final model. After following this procedure with each of our environmental variables, the AIC value was calculated for the model containing all remaining variables.
The final models were then validated by verifying the residuals of the GAMs for normality, homogeneity and independence. For each multivariate GAM, the distribution of residuals all fit the normal distribution. The heterogeneity for residuals was also checked, and there was no obvious pattern in the residuals against the fitted values. The over-dispersion parameter, which is computed by the sum of squared Pearson residuals divided by the sample size minus the number of parameters (Hilbe, 2011), was also applied to verify the over-dispersion of residuals (usually ranging from 0.8 to 1.5 (Zuur, 2012)). Inspection of the residuals for the longline GAMs showed temporal autocorrelation ranging from 0.2 to 0.4 at a distance of one time step (1 month). To remove the temporal autocorrelation an autoregressive model was added to the longline GAMs, such that all longline models are Generalized Additive Models with Auto Regres- sion (GAMAR) (Yang, Qin, Zhao, Wang, & Song, 2012). For the sake of simplicity, throughout the remainder of our paper we refer to both the GAM and GAMAR models as “GAM”. Table 2 contains some information on our multivariate GAMs’ distribution and validation."  
"Table 5 summarizes the performance for both periods of the multivariate GAMs for the CPUE of bigeye, yellowfin and skipjack by the purse seine and longline fishery. The table includes information on the explained deviance of model prediction (% Dev), its AIC, the degrees of freedom (df) or the estimated degrees of freedom (edf) for each of the variables, the p value of the significance of an envi- ronmental variable’s contribution to model prediction, and value of AIC for a model that lacks a given environmental variable (AIC’)."  
"Figure 6 (pre-FAD period) and Figure 7 (FAD period) show the environmental smooths of the GAM predictions of monthly biomass catch per set (CPS) of the purse seine fleet or monthly numerical catch per hook (CPH) of the longline fleet. Each figure displays a solid black line of the smooths, two dotted black lines of the upper and lower bounds of the 95% confidence limit of the smooth, and a dotted red line used to determine whether the values of the smooth are either above the lower confidence limit (indicating significantly higher CPUE) or below the upper confidence limit (indicating significantly lower CPUE). We summarize results in Table 6, and below."
- Arrizabalaga et al. (2015):  
"Scatterplots between environmental variables suggested high collinearity (R2 4 abs(0.6)) between TEMP5 and TEMP100, TEMP5 and O2, and DCHL and ZOO (Fig. 1). Thus, TEMP100, O2 and ZOO where excluded from the full GAMs. A stepwise backward model selection procedure was adopted, and the model with the lowest AIC was selected, following Chambers and Hastie (1992). Following the hierarchy principle, if an interaction was included in the model, lower order interactions and fixed factors were also included. The GAMs were built using the R v. 3.0 (R Development Core Team, 2013) and the mgcv 1.7–22 package (Wood, 2006). Missing values were omitted before the stepwise selection, so that model comparisons were based on the same dataset. The deviance explained by the model was estimated as 1-(residual deviance)/(null deviance), and is the equivalent to R2 in regression models (Guisan and Zimmermann, 2000)."  
"The full GAMs explained between 45% (for bigeye) and 64% (for skipjack) of the null deviance (Table 2). For all species, the full GAMs selected with the AIC criteria included all the 5 environmental variables TEMP5, SAL, SSH, MLD and DCHL. They also generally included the three way stock:year:season interaction for fixed factors, except in the case of albacore where the interactions stock:year and stock:season were kept in the final model, and in the case of southern bluefin, with a single stock, where the year: season interaction was included."  
"Table 2: Results of full GAMs, with final Akaike Information Criteria (AIC) and the percent deviance explained by the model (% Dev). For each variable retained in the final model, “df/ edf” indicates the degrees of freedom (for fixed factors) or estimated degrees of freedom (for environmental variables). Empty cells indicate that the variable was not included in the final model. AIC’ indicates the AIC value of the model when that particular variable was dropped during model selection. Dropping fixed factors or two way interactions was not tested if a higher order interaction was included in the model."

```python
for i, term in enumerate(gam.terms):
    if term.isintercept:
        continue
    XX = gam.generate_X_grid(term=i)
    pdep, confi = gam.partial_dependence(term=i, X=XX, width=0.95)
    plt.figure()
    plt.plot(XX[:, term.feature], pdep)
    plt.plot(XX[:, term.feature], confi, c='r', ls='--')
    plt.title(repr(term))
```
