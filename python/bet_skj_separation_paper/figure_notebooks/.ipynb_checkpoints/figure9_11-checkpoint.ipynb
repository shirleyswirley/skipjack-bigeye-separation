### Compute correl/regress coeff maps

```python
alphafdr = 0.1

ivnamelist = ['sst','temp100','tcd','sss',
              'o2100','po2100','thd',
              'logchl','mld','ssha']

bettoskjvsiv_cp_tot_rc = [None]*len(ivnamelist) 
bettoskjvsiv_cp_tot_rc_pval = [None]*len(ivnamelist) 
bettoskjvsiv_cp_tot_rc_ptf = [None]*len(ivnamelist)
bettoskjvsiv_cp_tot_cc = [None]*len(ivnamelist) 
bettoskjvsiv_cp_tot_cc_pval = [None]*len(ivnamelist) 
bettoskjvsiv_cp_tot_cc_ptf = [None]*len(ivnamelist)

betvsiv_cp_tot_rc = [None]*len(ivnamelist) 
betvsiv_cp_tot_rc_pval = [None]*len(ivnamelist) 
betvsiv_cp_tot_rc_ptf = [None]*len(ivnamelist)
betvsiv_cp_tot_cc = [None]*len(ivnamelist) 
betvsiv_cp_tot_cc_pval = [None]*len(ivnamelist) 
betvsiv_cp_tot_cc_ptf = [None]*len(ivnamelist)

skjvsiv_cp_tot_rc = [None]*len(ivnamelist) 
skjvsiv_cp_tot_rc_pval = [None]*len(ivnamelist) 
skjvsiv_cp_tot_rc_ptf = [None]*len(ivnamelist)
skjvsiv_cp_tot_cc = [None]*len(ivnamelist) 
skjvsiv_cp_tot_cc_pval = [None]*len(ivnamelist) 
skjvsiv_cp_tot_cc_ptf = [None]*len(ivnamelist)

for iiv in range(0,len(ivnamelist)):
    [bettoskjvsiv_cp_tot_rc[iiv], _, bettoskjvsiv_cp_tot_rc_pval[iiv],
     bettoskjvsiv_cp_tot_cc[iiv], bettoskjvsiv_cp_tot_cc_pval[iiv], _
    ]=gettemprcmap_loop(dsallvars[ivnamelist[iiv]],
                        dsallvars['bettoskj_cp_tot'],
                        'bettoskjvs'+ivnamelist[iiv])
    bettoskjvsiv_cp_tot_rc_ptf[iiv] = controlfdr2d(bettoskjvsiv_cp_tot_rc_pval[iiv],alphafdr)
    bettoskjvsiv_cp_tot_cc_ptf[iiv] = controlfdr2d(bettoskjvsiv_cp_tot_cc_pval[iiv],alphafdr)
    
    [betvsiv_cp_tot_rc[iiv], _, betvsiv_cp_tot_rc_pval[iiv],
     betvsiv_cp_tot_cc[iiv], betvsiv_cp_tot_cc_pval[iiv], _
    ]=gettemprcmap_loop(dsallvars[ivnamelist[iiv]],
                        dsallvars['bet_cp_tot'],
                        'betvs'+ivnamelist[iiv])
    betvsiv_cp_tot_rc_ptf[iiv] = controlfdr2d(betvsiv_cp_tot_rc_pval[iiv],alphafdr)
    betvsiv_cp_tot_cc_ptf[iiv] = controlfdr2d(betvsiv_cp_tot_cc_pval[iiv],alphafdr)
    
    [skjvsiv_cp_tot_rc[iiv], _, skjvsiv_cp_tot_rc_pval[iiv],
     skjvsiv_cp_tot_cc[iiv], skjvsiv_cp_tot_cc_pval[iiv], _
    ]=gettemprcmap_loop(dsallvars[ivnamelist[iiv]],
                        dsallvars['skj_cp_tot'],
                        'skjvs'+ivnamelist[iiv])
    skjvsiv_cp_tot_rc_ptf[iiv] = controlfdr2d(skjvsiv_cp_tot_rc_pval[iiv],alphafdr)
    skjvsiv_cp_tot_cc_ptf[iiv] = controlfdr2d(skjvsiv_cp_tot_cc_pval[iiv],alphafdr)
```
