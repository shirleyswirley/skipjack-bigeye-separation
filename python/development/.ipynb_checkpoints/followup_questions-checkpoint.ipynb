# Look at WCPFC data amts

### Plot # of sets

```python
unitsnow = '[# of sets]'
setslist = [wf.sets_una, wf.sets_log, wf.sets_dfad, wf.sets_afad, wf.sets_oth] 
setstitlelist = ['Unassociated sets', 'Natural log/debris sets', 'Drifting FAD sets', 'Anchored FAD sets', 'Other sets'] 

# - Set proj and define axes
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(16,7.5), subplot_kw={'projection': ccrs.PlateCarree(central_longitude=200)})

# - Make maps pretty + plot
isp = 0
for irow in range(axes.shape[0]):
    for icol in range(axes.shape[1]):
        if isp > len(setslist)-1:
            #axes[irow,icol].axis('off')
            axes[irow,icol].set_visible(False)
        else:
            ax = axes[irow][icol]
            exec(open("create_map_bgs.py").read())
            setslist[isp].sum(dim='time').plot(
                ax=ax, transform=ccrs.PlateCarree(), cmap=cm.viridis, #vmin=0, vmax=120000,
                cbar_kwargs={'pad': 0.02, 'label': unitsnow}) #, 'extend': 'max'})
            ax.set_xlabel(''); ax.set_ylabel('')
            ax.set_title(setstitlelist[isp])
            isp = isp + 1
        
fig.savefig(figpath + 'tot_number_diff_set_types.png', bbox_inches='tight', pad_inches = 0, dpi = 200)
```

```python
%run load_env_vars_orig.ipynb
```

```python
btosds = bettoskj_cp_tot.to_dataset(name='b2scptot')
btosds['skjcptot'] = skj_cp_tot
btosds['betcptot'] = bet_cp_tot
btosds['sst'] = sst
btosds['temp100'] = temp100
btosds['sss'] = sss
btosds['o2100'] = o2100
btosds['po2100'] = po2100
btosds['mld'] = mld
btosds['ssha'] = ssha
btosds['logchl'] = logchl
btosds['thd'] = thd
btosds['tcd'] = tcd
df = btosds.to_dataframe()
df.drop(columns=['depth'], inplace=True)
```

### Plot # of months of data

```python
# - Set proj and define axes
fig,axes = plt.subplots(nrows=4, ncols=3, figsize=(16,15), subplot_kw={'projection': ccrs.PlateCarree(central_longitude=200)})
colsnow = list(df.columns.values)
colsnow = colsnow[0:2] + colsnow[3:]
titlenow = ['BET:SKJ CPUE', 'SKJ CPUE', 'SST', 'Temp at 100m', 'Sea surface salinity',
            'O_2 at 100m', 'PO_2 at 100m', 'MLD', 'SSH anomalies', 'Chlorophyll a',
            'Tuna hypoxic depth', 'Thermocline depth']
unitsnow = '[# of months of data]'
ivar = 0
# - Make maps pretty + plot
for irow in range(axes.shape[0]):
    for icol in range(axes.shape[1]):
        ax = axes[irow][icol]
        exec(open("create_map_bgs.py").read())
        btosds[colsnow[ivar]].count(dim='time').plot(
            ax=ax, transform=ccrs.PlateCarree(),
            cmap=cm.viridis, vmin=0, vmax=610,
            cbar_kwargs={'pad': 0.02, 'label': unitsnow})
        ax.set_title(titlenow[ivar]); ax.set_xlabel(''); ax.set_ylabel('');
        ivar = ivar + 1 
fig.savefig(figpath + 'catch_and_env_condition_data_amts.png', bbox_inches='tight', pad_inches = 0, dpi = 200)
```

### Plot # of months of data overlapping w/ availability of BET:SKJ CPUE data

```python
# - Set proj and define axes
fig,axes = plt.subplots(nrows=4, ncols=3, figsize=(16,15), subplot_kw={'projection': ccrs.PlateCarree(central_longitude=200)})
colsnow = list(df.columns.values)
colsnow = colsnow[1:]
titlenow = ['SKJ CPUE', 'BET CPUE', 'SST', 'Temp at 100m', 'Sea surface salinity',
            'O_2 at 100m', 'PO_2 at 100m', 'MLD', 'SSH anomalies', 'Chlorophyll a',
            'Tuna hypoxic depth', 'Thermocline depth']
unitsnow = '[# of months of data]'
ivar = 0
# - Make maps pretty + plot
for irow in range(axes.shape[0]):
    for icol in range(axes.shape[1]):
        ax = axes[irow][icol]
        exec(open("create_map_bgs.py").read())
        mapnow = xr.where(btosds['b2scptot'].notnull() & btosds[colsnow[ivar]].notnull(), 1, 0)
        mapnow.sum(dim='time').plot(
            ax=ax, transform=ccrs.PlateCarree(),
            cmap=cm.viridis, vmin=0, vmax=520,
            cbar_kwargs={'pad': 0.02, 'label': unitsnow})
        ax.set_title(titlenow[ivar]); ax.set_xlabel(''); ax.set_ylabel('');
        ivar = ivar + 1 
fig.savefig(figpath + 'b2scpue_env_condition_overlapping_data_amts.png', bbox_inches='tight', pad_inches = 0, dpi = 200)
```
