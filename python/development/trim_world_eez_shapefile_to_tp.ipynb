```python
import geopandas as gpd
import pandas as pd
```

```python
# - Set data path
dpath = '/opt/skipjack-bigeye-separation/data/'
```

```python
eezs = gpd.read_file(
    dpath + 'EEZs/World_EEZ_v9_20161021_HR_0_360/World_EEZ_v9_2016_HR_0_360.shp')
```

```python
# Note that Sovereign1 = country name
nuctries = ['Micronesia','Kiribati','Marshall Islands','Nauru','Palau','Papua New Guinea','Solomon Islands','Tuvalu'];
etpctries = ['Peru','Ecuador','Colombia','Panama','Costa Rica','Nicaragua','El Salvador','Guatemala'];
```

```python
eezstp = eezs[eezs['Sovereign1'].isin(nuctries+etpctries)]
```

```python
eezstp = eezstp[['PolygonID','GeoName','Sovereign1','geometry']]
```
