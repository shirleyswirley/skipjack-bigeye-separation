```python
# - Define constant plot params
stipsizenow = 10
stipmarknow = 'o'
stipfacecolnow = 'none'
stipedgeltcolnow = 'whitesmoke'
stipewnow = 0.8 # marker edge width
eezfcnow = 'none'
eezlcnow = 'silver'
eezlsnow = '-'
eezlwnow = 1.5
```

```python
# - Define subplot-variable plot params
if mapvarnow=='skj':
    fignamenow = 'suppfig4'
    unitsnow = 4*['[metric tons/set]']
    mapsnow = [skj_cp_tot_en, skj_cp_tot_en-skj_cp_tot_mean,
               skj_cp_tot_ln, skj_cp_tot_ln-skj_cp_tot_mean]
    vmaxsnow = [60, 12, 60, 12]
    pvsnow = 4*[skj_cp_tot_enln_wrs_pval]
    ptfsnow = 4*[skj_cp_tot_enln_wrs_ptf]
    titlesnow = ['SKJ CPUE - El Niño', 'SKJ CPUE - El Niño minus mean',
                 'SKJ CPUE - La Niña', 'SKJ CPUE - La Niña minus mean']
    vminsnow = [0, -12, 0, -12]
elif mapvarnow=='bet':
    fignamenow = 'suppfig5'
    unitsnow = 4*['[metric tons/set]']
    mapsnow = [bet_cp_tot_en, bet_cp_tot_en-bet_cp_tot_mean,
               bet_cp_tot_ln, bet_cp_tot_ln-bet_cp_tot_mean]
    vmaxsnow = [15, 4, 15, 4]
    pvsnow = 4*[bet_cp_tot_enln_wrs_pval]
    ptfsnow = 4*[bet_cp_tot_enln_wrs_ptf]
    titlesnow = ['BET CPUE - El Niño', 'BET CPUE - El Niño minus mean',
                 'BET CPUE - La Niña', 'BET CPUE - La Niña minus mean']
    vminsnow = [0, -4, 0, -4]
elif mapvarnow=='bettoskj':
    fignamenow = 'mainfig4'
    unitsnow = 4*['']
    mapsnow = [bettoskj_cp_tot_en, bettoskj_cp_tot_en-bettoskj_cp_tot_mean,
               bettoskj_cp_tot_ln, bettoskj_cp_tot_ln-bettoskj_cp_tot_mean]
    vmaxsnow = [0.32, 0.12, 0.32, 0.12]
    pvsnow = 4*[bettoskj_cp_tot_enln_wrs_pval]
    ptfsnow = 4*[bettoskj_cp_tot_enln_wrs_ptf]
    titlesnow = ['BET:SKJ catch - El Niño', 'BET:SKJ catch - El Niño minus mean',
                 'BET:SKJ catch - La Niña', 'BET:SKJ catch - La Niña minus mean']
    vminsnow = [0, -0.12, 0, -0.12]

cmseqnow = plt.cm.get_cmap('viridis',11)
cmdivnow = plt.cm.get_cmap('PuOr',11)
cmsnow = [cmseqnow, cmdivnow,
          cmseqnow, cmdivnow]
stipedgedkcolsnow = ['lightgray','black',
                     'lightgray','black']
signifstipsnow = [0,1,0,1]
ploteezsnow = [1,0,1,0]
stipltdkcosnow = 0.5*np.asarray(vmaxsnow) # light/dark stip cutoff value
```

```python
# - Set proj and define axes
fig,axes = plt.subplots(nrows=2, ncols=2, figsize=(10,4),
                        subplot_kw={'projection': ccrs.PlateCarree(central_longitude=200)})

# - Make maps pretty + plot
isp = 0
for irow in range(axes.shape[0]):
    for icol in range(axes.shape[1]):
        ax = axes[irow][icol]
        exec(open('helper_scripts/create_map_bgs.py').read())
        
        mapsnow[isp].plot(
            ax=ax, transform=ccrs.PlateCarree(), cmap=cmsnow[isp],
            vmin=vminsnow[isp], vmax=vmaxsnow[isp],
            cbar_kwargs={'pad': 0.02, 'label': unitsnow[isp]})
        
        if ploteezsnow[isp]==1:
            nueezs.plot(ax=ax, transform=ccrs.PlateCarree(),
                color=eezfcnow, edgecolor=eezlcnow, linewidth=eezlwnow)
            
        if signifstipsnow[isp]==1:
            [ltcol_signiflonnow,ltcol_signiflatnow]=find_where_pval_small(
                pvsnow[isp].where(abs(mapsnow[isp])>stipltdkcosnow[isp]),
                ptfsnow[isp])
            [dkcol_signiflonnow,dkcol_signiflatnow]=find_where_pval_small(
                pvsnow[isp].where(abs(mapsnow[isp])<=stipltdkcosnow[isp]),
                ptfsnow[isp])
            ax.scatter(ltcol_signiflonnow, ltcol_signiflatnow,
                       marker=stipmarknow, linewidths=stipewnow,
                       facecolors=stipfacecolnow, edgecolors=stipedgeltcolnow,
                       s=stipsizenow, transform=ccrs.PlateCarree())
            ax.scatter(dkcol_signiflonnow, dkcol_signiflatnow,
                       marker=stipmarknow, linewidths=stipewnow,
                       facecolors=stipfacecolnow, edgecolors=stipedgedkcolsnow[isp],
                       s=stipsizenow, transform=ccrs.PlateCarree())
                    
        ax.set_xlabel(''); ax.set_ylabel('')
        ax.set_title(titlesnow[isp])
        isp = isp + 1
        
fig.savefig(figpath + fignamenow + '_bscpue_enso_maps.png',
            bbox_inches='tight', pad_inches = 0, dpi = 200)       
```
