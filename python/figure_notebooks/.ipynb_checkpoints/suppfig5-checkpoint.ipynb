```python
# - Define subplot-variable plot params
unitsnow = 12*['']
mapsnow = [bettoskj_cp_tot_seas.sel(season='DJF'),
           bettoskj_cp_tot_seas.sel(season='DJF')-bettoskj_cp_tot_mean,
           summary_map_win, 
           bettoskj_cp_tot_seas.sel(season='MAM'),
           bettoskj_cp_tot_seas.sel(season='MAM')-bettoskj_cp_tot_mean,
           summary_map_spr, 
           bettoskj_cp_tot_seas.sel(season='JJA'),
           bettoskj_cp_tot_seas.sel(season='JJA')-bettoskj_cp_tot_mean,
           summary_map_sum, 
           bettoskj_cp_tot_seas.sel(season='SON'),
           bettoskj_cp_tot_seas.sel(season='SON')-bettoskj_cp_tot_mean,
           summary_map_aut]
hatchmapsnow = [None,None,bettoskj_cp_tot_seas.sel(season='DJF'),
                None,None,bettoskj_cp_tot_seas.sel(season='MAM'),
                None,None,bettoskj_cp_tot_seas.sel(season='JJA'),
                None,None,bettoskj_cp_tot_seas.sel(season='SON')]
vmaxsnow = 4*[0.32, 0.12, 0.25]
vminsnow = 4*[0, -0.12, 5.25]

pvsnow = 12*[bettoskj_cp_tot_seas_kw_pval]
ptfsnow = 12*[bettoskj_cp_tot_seas_kw_ptf]

cmseqnow = plt.cm.get_cmap('viridis',11)
cmdivnow = plt.cm.get_cmap('PuOr',11)
cmsnow = 4*[cmseqnow, cmdivnow, None]

stipltdkcosnow = 0.5*np.asarray(vmaxsnow) # light/dark stip cutoff value
stipedgedkcolsnow = 4*['lightgray', 'darkslategray', None]
signifstipsnow = 4*[0, 1, None]
ploteezsnow = 4*[1, 0, None]

titlesnow = ['BET:SKJ - Winter','BET:SKJ - Winter minus mean','Winter summary',
             'BET:SKJ - Spring','BET:SKJ - Spring minus mean','Spring summary',
             'BET:SKJ - Summer','BET:SKJ - Summer minus mean','Summer summary',
             'BET:SKJ - Fall','BET:SKJ - Fall minus mean','Fall summary']
```
