```python
# - Define constant plot params
stipsizenow = 10; stipmarknow = 'o'
stipfacecolnow = 'none'
stipedgeltcolnow = 'whitesmoke'
stipewnow = 0.8 # marker edge width

eezfcnow = 'none'; eezlcnow = 'lightgray' #'silver'
eezlsnow = '-'; eezlwnow = 0.9 #1.5

bsstipsizenow = 50 #100
bsstipfacecolnow = 'white'
bsstipedgecolnow = 'white'
bsstipewnow = 0.5 # marker edge width

bseezfcnow = 'none'; bseezlcnow = 'black';
bseezlsnow = '-'; bseezlwnow = 0.8
```

```python
# - Define subplot-variable plot params
fignamenow = 'mainfig4'
unitsnow = 4*['']
mapsnow = [bettoskj_cp_tot_en, bettoskj_cp_tot_ln,
           bettoskj_cp_tot_en-bettoskj_cp_tot_mean,
           bettoskj_cp_tot_ln-bettoskj_cp_tot_mean]
vmaxsnow = [0.32, 0.32, 0.12, 0.12]
vminsnow = [0, 0, -0.12, -0.12]
pvsnow = 4*[bettoskj_cp_tot_enln_wrs_pval]
ptfsnow = 4*[bettoskj_cp_tot_enln_wrs_ptf]
cmseqnow = plt.cm.get_cmap('viridis',11)
cmdivnow = plt.cm.get_cmap('PuOr',11)
cmsnow = [cmseqnow, cmseqnow, cmdivnow, cmdivnow]
stipltdkcosnow = 0.5*np.asarray(vmaxsnow) # light/dark stip cutoff value
stipedgedkcolsnow = ['lightgray','lightgray','black','black']
signifstipsnow = [0,0,1,1]
ploteezsnow = [1,1,0,0]
titlesnow = ['BET:SKJ - El Niño', 'BET:SKJ - La Niña',
             'BET:SKJ - El Niño minus mean',
             'BET:SKJ - La Niña minus mean', 
             'El Niño summary', 'La Niña summary']
```

```python
# - Set proj and define axes
fig,axes = plt.subplots(nrows=3, ncols=2, figsize=(7.5,5.3),
                        subplot_kw={'projection': ccrs.PlateCarree(central_longitude=200)})

# - Make maps pretty + label subplots + plot
isp = 0
for irow in range(axes.shape[0]):
    for icol in range(axes.shape[1]):
        ax = axes[irow][icol]
        exec(open('helper_scripts/create_map_bgs.py').read())
        ax.text(-0.1, 1.1, string.ascii_uppercase[isp], transform=ax.transAxes,
                size=12, weight='bold') 
        
        if isp<4:
            pltnow = mapsnow[isp].plot(add_colorbar=False,
                ax=ax, transform=ccrs.PlateCarree(), cmap=cmsnow[isp],
                vmin=vminsnow[isp], vmax=vmaxsnow[isp])
            cbar = plt.colorbar(pltnow, ax=ax,
                                pad=0.02, label=unitsnow[isp], shrink=0.9)
            cbar.ax.tick_params(labelsize=8)
            
            if ploteezsnow[isp]==1:
                nueezs.plot(ax=ax, transform=ccrs.PlateCarree(),
                    color=eezfcnow, edgecolor=eezlcnow,
                    linewidth=eezlwnow, linestyle=eezlsnow)
                
            if signifstipsnow[isp]==1:
                [ltcol_signiflonnow,ltcol_signiflatnow]=find_where_pval_small(
                    pvsnow[isp].where(abs(mapsnow[isp])>stipltdkcosnow[isp]),
                    ptfsnow[isp])
                [dkcol_signiflonnow,dkcol_signiflatnow]=find_where_pval_small(
                    pvsnow[isp].where(abs(mapsnow[isp])<=stipltdkcosnow[isp]),
                    ptfsnow[isp])
                ax.scatter(ltcol_signiflonnow, ltcol_signiflatnow,
                           marker=stipmarknow, linewidths=stipewnow,
                           facecolors=stipfacecolnow, edgecolors=stipedgeltcolnow,
                           s=stipsizenow, transform=ccrs.PlateCarree())
                ax.scatter(dkcol_signiflonnow, dkcol_signiflatnow,
                           marker=stipmarknow, linewidths=stipewnow,
                           facecolors=stipfacecolnow, edgecolors=stipedgedkcolsnow[isp],
                           s=stipsizenow, transform=ccrs.PlateCarree())
                    
        if isp>=4:
            if isp==4:
                summarymapnow = summary_map_en
                hatchmapnow = bettoskj_cp_tot_en 
            elif isp==5:
                summarymapnow = summary_map_ln
                hatchmapnow = bettoskj_cp_tot_ln 
            hatchmapco1now = bettoskj_r3co1 
            hatchmapco2now = bettoskj_r3co2
            cbarlabelsnow = ['LS-LB','LS-HB','HS-LB','MS-MB','HS-HB']
            
            pltnow = summarymapnow.plot(ax=ax, add_colorbar=False,
                                   transform=ccrs.PlateCarree(),
                                   cmap=bs5cmp_nonone, vmin=0.25, vmax=5.25)
            
            hmap = hatchmapnow.where(hatchmapnow>=hatchmapco2now)
            hmap_stacked = hmap.stack(x=['lat','lon'])
            hmap_red = hmap_stacked[hmap_stacked.notnull()]
            bsstipmarknow = 'x'
            ax.scatter(hmap_red.lon, hmap_red.lat,
                        marker=bsstipmarknow, linewidths=bsstipewnow,
                        facecolors=bsstipfacecolnow, edgecolors=bsstipedgecolnow,
                        s=bsstipsizenow, transform=ccrs.PlateCarree())
            
            hmap = hatchmapnow.where((hatchmapnow<hatchmapco2now)
                                     &(hatchmapnow>hatchmapco1now))
            hmap_stacked = hmap.stack(x=['lat','lon'])
            hmap_yellow = hmap_stacked[hmap_stacked.notnull()]
            bsstipmarknow = 'None' #'$M$'
            ax.scatter(hmap_yellow.lon, hmap_yellow.lat,
                        marker=bsstipmarknow, linewidths=bsstipewnow,
                        facecolors=bsstipfacecolnow, edgecolors=bsstipedgecolnow,
                        s=bsstipsizenow, transform=ccrs.PlateCarree())
            
            hmap = hatchmapnow.where(hatchmapnow<=hatchmapco1now)
            hmap_stacked = hmap.stack(x=['lat','lon'])
            hmap_green = hmap_stacked[hmap_stacked.notnull()]
            bsstipmarknow = '$\checkmark$'
            ax.scatter(hmap_green.lon, hmap_green.lat,
                        marker=bsstipmarknow, linewidths=bsstipewnow,
                        facecolors=bsstipfacecolnow, edgecolors=bsstipedgecolnow,
                        s=bsstipsizenow, transform=ccrs.PlateCarree())
            
            cbar = plt.colorbar(pltnow, ax=ax, ticks=np.arange(0.25,5.25,1) + 1/2,
                                pad=0.02, label='', shrink=0.9)
            cbar.ax.set_yticklabels(cbarlabelsnow)
            cbar.ax.tick_params(labelsize=8)
            
            nueezs.plot(ax=ax, transform=ccrs.PlateCarree(),
                        color=bseezfcnow, edgecolor=bseezlcnow,
                        linewidth=bseezlwnow, linestyle=bseezlsnow)

        ax.set_xlabel(''); ax.set_ylabel('')
        ax.tick_params(axis='x', labelsize=8)
        ax.tick_params(axis='y', labelsize=8)
        ax.set_title(titlesnow[isp], fontsize=10)
        isp = isp + 1

# - Save fig
fig.savefig(figpath + 'Fig4.png',
            bbox_inches='tight', pad_inches = 0, dpi = 300)       
fig.savefig(figpath + 'Fig4.tif',
            bbox_inches='tight', pad_inches = 0, dpi = 300,
            pil_kwargs={"compression": "tiff_lzw"})
```
